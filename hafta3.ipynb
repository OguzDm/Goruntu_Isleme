```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def getJpegFiles():
    os.getcwd()
    os.listdir()
    path = os.getcwd()
    jpg_files = [f for f in os.listdir(path) if f.endswith('.jpg')]
    return jpg_files

jpgFiles = getJpegFiles()
```

```python
catImage = plt.imread(jpgFiles[2])
```

```python
def get_value_from_triple(temp1):
    return int(temp1[0]/3+temp1[1]/3+temp1[2]/3)
get_value_from_triple(catImage[0,0,:])


def get_0_1_from_triple(temp_1):
    mean = int(temp_1[0]/3 + temp_1[1]/3 + temp_1[2]/3)
    if mean < 110:
        return 0
    else:
        return 1
```

```python
def convert_rgb_to_gray(catImage):
    m,n,k = catImage.shape
    newImage = np.zeros((m,n),dtype ='uint8')

    for i in range(m):
        for j in range(n):
            s = get_value_from_triple(catImage[i,j,:])
            newImage[i,j] = s
        
    return newImage

catImage2 = convert_rgb_to_gray(catImage)

def convert_rgb_to_bw(catImage):
    m,n,k = catImage.shape
    newImage = np.zeros((m,n),dtype ='uint8')

    for i in range(m):
        for j in range(n):
            s = get_0_1_from_triple(catImage[i,j,:])
            newImage[i,j] = s
        
    return newImage

catImage3 = convert_rgb_to_bw(catImage)
```
